Importing dependencies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Loading the Boston House Price Dataset

```python
house_dataset = pd.read_csv("Boston-house-price-data.csv")
house_dataset.head()
```

```python
house_dataset.shape
```

```python
house_dataset.isnull().sum()
```

```python
house_dataset.info()
```

```python
house_dataset.describe()
```

Understand the Correlation btw different features in the dataset              
1.Positive correlation
2.Negative correlation


```python
correlate = house_dataset.corr()
correlate
```

Construct a Heatmap to understand the Correlation                               
here, "MEDV" refers to " Price "

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlate, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
X = house_dataset.drop(['MEDV'], axis=1)
Y = house_dataset['MEDV']
```

```python
display(X)
display(Y)
```

Splitting the data into training data and splitting data

```python
X_train ,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training


XGBoost Regression

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

Visualizing the Actual vs Predicted values

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
